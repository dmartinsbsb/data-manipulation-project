```python
# === Bibliotecas =========
import pandas as pd
```

```python
"""
O time de marketing quer segmentar os clientes com base em sua faixa etária e padrão de compras. Eles pedem:

Um arquivo contendo:

ID do Cliente

Faixa Etária (por exemplo: "<25", "25-40", ">40")

Média de gastos por cliente

Número total de compras realizadas

Adicione uma coluna categorizando os clientes como "Alta Frequência" (mais de 20 compras) ou "Baixa Frequência".

"""
```

```python
#Lendo os dados
df = pd.read_csv("../data/processed/dataset_processed.csv")
```

```python
df.head() # Verificando se tudo foi carregado corretamente
```

```python
# Criando uma função que classifica a faixa etária 

def faixa_etaria(idade):
    if idade < 25:
        return "<25"
    elif 25 <= idade <= 40:
        return "25-40"
    else:
        return ">40"


# Adicionar a coluna Faixa Etária
df['Faixa_Etaria'] = df['Idade'].apply(faixa_etaria)
```

```python
df.head()
```

```python
# Criando a coluna de média de gastos por cliente
df["Media_Gastos"] = df.groupby("ID_Cliente")["Valor_Compra"].transform("mean")
df.head() # visualizando para ver se a coluna foi adicionada
```

```python
# Criando a coluna de número total de compras realizadas
df["Num_Compras"] = df.groupby("ID_Cliente")["Valor_Compra"].transform("count")
df.head() # visualizando para ver se a coluna foi adicionada
```

```python
# Função para classificar a frequência de compras
def frequencia_compras(num_compras):
    if num_compras > 20:
        return 'Alta Frequência'
    else:
        return 'Baixa Frequência'

# Adicionar a coluna Frequência de Compras
df['Frequencia_Compras'] = df['Num_Compras'].apply(frequencia_compras)
df.head()
```

```python
df_selecionado = df[['ID_Cliente', 'Faixa_Etaria', 'Media_Gastos', 'Num_Compras', 'Frequencia_Compras']] # Selecionando apenas as colunas que queremos
```

```python
df_selecionado = df_selecionado.drop_duplicates(subset="ID_Cliente") # Tirando os dados duplicados
```

```python
df_selecionado.head(10) # df final
```

```python
df_selecionado.to_csv("../data/processed/marketing_request.csv", index=False)
```
